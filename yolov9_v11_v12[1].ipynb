```python
from ultralytics.utils.metrics import ConfusionMatrix
from pathlib import Path
import numpy as np
import os

gt_path = Path('/kaggle/working/clc-xray-yolo/test/labels')
pred_path = Path('/kaggle/working/runs/detect/predict/labels')

gt_files = sorted(list(gt_path.glob('*.txt')))
pred_files = sorted(list(pred_path.glob('*.txt')))

confmat = ConfusionMatrix(nc=1)

def read_yolo_file(file_path):
    if not os.path.exists(file_path):
        return np.zeros((0, 5))
    return np.loadtxt(file_path, ndmin=2)

for gt_file in gt_files:
    image_id = gt_file.stem
    pred_file = pred_path / f"{image_id}.txt"

    gt_boxes = read_yolo_file(gt_file)
    pred_boxes = read_yolo_file(pred_file)

    if gt_boxes.shape[1] == 5 and pred_boxes.shape[1] == 5:
        confs = np.ones((len(pred_boxes), 1))  # assuming confidence = 1
        pred_combined = np.concatenate((pred_boxes, confs), axis=1)
        confmat.process_batch(pred_combined, gt_boxes)

# Get the confusion matrix
cm = confmat.matrix  # shape: (nc+1, nc+1) ‚Äî includes background

# Extract TP, FP, FN
tp = np.diag(cm)[:-1]  # diagonal entries except last (background)
fp = cm[:-1, -1]       # predictions of class i as background
fn = cm[-1, :-1]       # background predicted as class i

# Calculate metrics
precision = tp / (tp + fp + 1e-6)
recall = tp / (tp + fn + 1e-6)
f1 = 2 * precision * recall / (precision + recall + 1e-6)

# Output
print(f"Confusion Matrix:\n{cm}")
print(f"\nPrecision: {precision[0]:.4f}")
print(f"Recall:    {recall[0]:.4f}")
print(f"F1 Score:  {f1[0]:.4f}")
```

```python
from ultralytics.utils.metrics import ConfusionMatrix
from pathlib import Path
import numpy as np
import os

# Class info
num_classes = 13
names = ['gun-knife-wrench-pilers-scissors', 'Cans', 'CartonDrinks', 'GlassBottle',
         'PlasticBottle', 'SprayCans', 'SwissArmyKnife', 'Tin', 'VacuumCup',
         'blade', 'dagger', 'knife', 'scissors']

# Paths
gt_path = Path('/kaggle/working/clc-xray-yolo/test/labels')
pred_path = Path('/kaggle/working/runs/detect/predict/labels')

confmat = ConfusionMatrix(nc=num_classes)

def read_yolo_file(path):
    if not path.exists() or os.path.getsize(path) == 0:
        return np.zeros((0, 5))
    try:
        return np.loadtxt(path, ndmin=2)
    except:
        return np.zeros((0, 5))

# Evaluate predictions
for pred_file in pred_path.glob('*.txt'):
    stem = pred_file.stem
    gt_file = gt_path / f"{stem}.txt"

    pred_boxes = read_yolo_file(pred_file)
    gt_boxes = read_yolo_file(gt_file)

    if pred_boxes.shape[1] == 5 and gt_boxes.shape[1] == 5:
        # Extract components
        pred_cls = pred_boxes[:, 0].astype(int)
        pred_xywh = pred_boxes[:, 1:]
        confs = np.ones((len(pred_boxes), 1))
        detections = np.concatenate((pred_xywh, confs), axis=1)

        gt_cls = gt_boxes[:, 0].astype(int)
        labels = gt_boxes[:, 1:]

        confmat.process_batch(detections, pred_cls, labels, gt_cls)

# Print confusion matrix
print("Confusion Matrix:\n", confmat.matrix)

# Metrics per class
for i in range(num_classes):
    tp = confmat.matrix[i, i]
    fp = confmat.matrix[:, i].sum() - tp
    fn = confmat.matrix[i, :].sum() - tp
    precision = tp / (tp + fp + 1e-6)
    recall = tp / (tp + fn + 1e-6)
    f1 = 2 * precision * recall / (precision + recall + 1e-6)

    print(f"\nClass {i} ({names[i]}):")
    print(f"  Precision: {precision:.4f}")
    print(f"  Recall:    {recall:.4f}")
    print(f"  F1 Score:  {f1:.4f}")
```
